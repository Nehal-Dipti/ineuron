```python
def profit_flag(a):
    if a>0:
        return 'positive'
    else:
        return 'negative'    
```

```python
df_sales['flag_profit'] = df_sales['profit'].apply(profit_flag)
```

```python
df_sales['len_cust_name'] = df_sales['customer_name'].apply(len)
```

```python
def quantity_falg(a):
    if a<10:
        return 'low'
    elif a>10 and a<20:    
        return 'medium'
    else :
        return 'high'    
```

```python
df_sales['flag_quantity']  = df_sales['quantity'].apply(quantity_falg)
```
